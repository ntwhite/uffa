```python
import uffa
import numpy as np
import pandas as pd
from random import seed
from random import random

```

```python
# generate the random walk
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
	movement = -1 if random() < 0.5 else 1
	value = random_walk[i-1] + movement
	random_walk.append(value)
```

```python
#INSTANTIATE CLASS
data_obj = uffa.uffa_stack(random_walk,100)
```

```python
data_obj.runner_func()
```

```python
#SPLIT DATA
train, test = data_obj.train_test_split()
```

```python
#PERSISTENCE MODEL RESULTS
pers_mode_results = data_obj.persistence_model(train,test)
```

```python
#RW MODEL RESULTS
el_rw_results = data_obj.elementary_rw(train,test)
```

```python
#RW + MEAN MODEL RESULTS
el_rw_rm = data_obj.elementary_rw_nm(train,test)
```

```python
#AUTO ARIMA MODEL RESULTS
auto_am_results = data_obj.auto_pmd(train,test)
```

```python
x = data_obj.report_creation(test,pers_mode_results, el_rw_results, el_rw_rm, auto_am_results)
print(x)
```
