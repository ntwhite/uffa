```python
import pandas as pd
import numpy as np
from random import seed
from random import random
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt, ExponentialSmoothing

import accuarcy_func as af
%matplotlib inline
```

```python
# generate the random walk
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
	movement = -1 if random() < 0.5 else 1
	value = random_walk[i-1] + movement
	random_walk.append(value)
# prepare dataset
train_size = int(len(random_walk) * 0.78)
train, test = random_walk[0:train_size], random_walk[train_size:]
```

```python
import matplotlib.pyplot as plt
plt.plot(train)
plt.ylabel('some numbers')
plt.show()
```

## Import auto-arima

```python
import pmdarima as aa
from pmdarima.arima import auto_arima
from pmdarima.arima.utils import ndiffs, nsdiffs
```

```python
little_d = ndiffs(train, test = 'kpss') 
big_D = nsdiffs(train, m = 52, max_D = 12, test = 'ocsb')

model_1 = aa.auto_arima(train, 
                        start_p = 0,
                        start_q = 0,
                        max_p = 5,
                        max_q = 5,
                        m = 52,
                        start_P = 0,
                        seasonal = True,
                        d = little_d,
                        D = big_D,
                        suppress_warnings = True,
                        stepwise = True,
                        error_action = 'ignore',
                        trace = True)
```

```python
fcast = model_1.predict(n_periods = 220)
```

```python
test = pd.Series(test)
test.index = range(780,1000)
fcast = pd.Series(fcast)
fcast.index = test.index

plt.plot(train, label = 'train')
plt.plot(test, label = 'test')
plt.plot(fcast, label = 'forecast')
plt.legend(loc = 'lower left')
```

## el random walk

```python
def elementary_rw(train,test):
    '''
    create an elementary rw model where step move sizes have an equal probability of moving up or down, and step size equals the standard deviation of the training set
    '''
    seed(1)
    stand_dev = np.std(train)
    
    start_value = train[-1]
    predictions = list()
        
    for i in test:
        movement = -stand_dev if random() < 0.5 else stand_dev
        value = start_value + movement
        start_value = value
        predictions.append(value)
    
    return predictions
            
    
    
```

```python
elementary_rw(train,test)
```

```python
def elementary_rw_nm(train,test):
    '''
    create an elementary rw model where step move sizes have an equal probability of moving up or down, and step size equals the standard deviation of the training set
    '''
    seed(1)
    stand_dev = np.std(train)
    
    start_value = train[-1]
    predictions = list()
        
    for i in test:
        num = random()
        if num <= 0.33:
            movement = stand_dev
        elif num <= 0.66:
            movement = -stand_dev
        else:
            movement = 0
        value = start_value + movement
        start_value = value
        predictions.append(value)
    
    return predictions
```

```python
elementary_rw_nm(train,test)
```
