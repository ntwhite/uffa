## Import Libs

```python
import uffa
import numpy as np
import pandas as pd
from random import seed
from random import random
import matplotlib.pyplot as plt
```

## Import/Create Data

```python
# generate the random walk seed 22
seed(22)
random_walk_22 = list()
random_walk_22.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
	movement = -1 if random() < 0.5 else 1
	value = random_walk_22[i-1] + movement
	random_walk_22.append(value)
    
    
seed(1)
random_walk_1 = list()
random_walk_1.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
	movement = -1 if random() < 0.5 else 1
	value = random_walk_1[i-1] + movement
	random_walk_1.append(value)


#Arline_pass data
ap_df = pd.read_csv('AirPassengers.csv')

#BBY Data 1
df_301 = pd.read_csv('df_301_computing.csv')

#BBY
df_302 = pd.read_csv('df_319_computing.csv')
```

## Random Walk Data Assessment (seed 22)

```python
#INSTANTIATE CLASS
data_obj = uffa.uffa_stack(random_walk_22,100)

#SPLIT DATA
train, test = data_obj.train_test_split()

#PERSISTENCE MODEL RESULTS
pers_mode_results = data_obj.persistence_model(train,test)

#RW MODEL RESULTS
el_rw_results = data_obj.elementary_rw(train,test)

#RW + MEAN MODEL RESULTS
el_rw_rm = data_obj.elementary_rw_nm(train,test)

#AUTO ARIMA MODEL RESULTS
auto_am_results = data_obj.auto_pmd(train,test)

#accuracy Creation
pers_acc = data_obj.all_acc(test,pers_mode_results)
el_rw_acc = data_obj.all_acc(test,el_rw_results)
el_rw_m_acc = data_obj.all_acc(test,el_rw_rm)
auto_am_acc = data_obj.all_acc(test,auto_am_results)

#### I NEED TO ADD ALL THIS STUFF TO MY CLASS
x = pd.concat([pers_acc, el_rw_acc, el_rw_m_acc, auto_am_acc])
x['model'] = ['persistence','random_walk','random_walk_m','auto_classical']
#x

test = pd.Series(test)
test.index = range(900,1000)
fcast = pd.Series(auto_am_results)
fcast.index = test.index

plt.plot(train, label = 'train')
plt.plot(test, label = 'test')
plt.plot(fcast, label = 'auto_am_results')
plt.legend(loc = 'lower left')
x
```
